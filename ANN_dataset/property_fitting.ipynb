As a first step following the supplementary information of this paper - we must find the temperature dependence of the electrical conductivity Seebeck coefficient and Thermal conductivity (https://www.sciencedirect.com/science/article/pii/S030626192101134X#b0265)

Using the last estm dataset actually it should be possible to fit equations to these 

```python
import pandas as pd
file_path = 'estm.xlsx'
data = pd.read_excel(file_path)
print(data.head(10))
```

```python
# Filter data for PbTe and SnSe
materials_of_interest = ['PbTe', 'SnSe']
filtered_data = data[data['Formula'].isin(materials_of_interest)]
print(filtered_data.head(100))
```

```python
import matplotlib.pyplot as plt
# Plot thermoelectric properties for PbTe
pbte_data = filtered_data[filtered_data['Formula'] == 'PbTe']

fig, axs = plt.subplots(3, 1, figsize=(8, 15), sharex=True)

# Seebeck coefficient
axs[0].scatter(pbte_data['temperature(K)'], pbte_data['seebeck_coefficient(μV/K)'], label='PbTe', color='blue')
axs[0].set_title('PbTe: Seebeck Coefficient vs Temperature')
axs[0].set_ylabel('Seebeck Coefficient (μV/K)')
axs[0].legend()

# Electrical conductivity
axs[1].scatter(pbte_data['temperature(K)'], pbte_data['electrical_conductivity(S/m)'], label='PbTe', color='green')
axs[1].set_title('PbTe: Electrical Conductivity vs Temperature')
axs[1].set_ylabel('Electrical Conductivity (S/m)')
axs[1].legend()

# Thermal conductivity
axs[2].scatter(pbte_data['temperature(K)'], pbte_data['thermal_conductivity(W/mK)'], label='PbTe', color='red')
axs[2].set_title('PbTe: Thermal Conductivity vs Temperature')
axs[2].set_xlabel('Temperature (K)')
axs[2].set_ylabel('Thermal Conductivity (W/mK)')
axs[2].legend()

plt.tight_layout()
plt.show()

# Plot thermoelectric properties for SnSe
snse_data = filtered_data[filtered_data['Formula'] == 'SnSe']

fig, axs = plt.subplots(3, 1, figsize=(8, 15), sharex=True)

# Seebeck coefficient
axs[0].scatter(snse_data['temperature(K)'], snse_data['seebeck_coefficient(μV/K)'], label='SnSe', color='blue')
axs[0].set_title('SnSe: Seebeck Coefficient vs Temperature')
axs[0].set_ylabel('Seebeck Coefficient (μV/K)')
axs[0].legend()

# Electrical conductivity
axs[1].scatter(snse_data['temperature(K)'], snse_data['electrical_conductivity(S/m)'], label='SnSe', color='green')
axs[1].set_title('SnSe: Electrical Conductivity vs Temperature')
axs[1].set_ylabel('Electrical Conductivity (S/m)')
axs[1].legend()

# Thermal conductivity
axs[2].scatter(snse_data['temperature(K)'], snse_data['thermal_conductivity(W/mK)'], label='SnSe', color='red')
axs[2].set_title('SnSe: Thermal Conductivity vs Temperature')
axs[2].set_xlabel('Temperature (K)')
axs[2].set_ylabel('Thermal Conductivity (W/mK)')
axs[2].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from numpy.polynomial.polynomial import Polynomial

# Load dataset
data = pd.read_excel('estm.xlsx')

# Define materials and references
pbte_data = data[data['Formula'] == 'PbTe']
snse_ref = 'https://doi.org/10.1016/j.jallcom.2016.01.190'
snse_data = data[(data['Formula'] == 'SnSe') & (data['reference'] == snse_ref)]

fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# Plotting function with polynomial fitting
def plot_with_fit(ax, x, y, degree, title, xlabel, ylabel):
    ax.scatter(x, y, label='Data')
    coefs = np.polyfit(x, y, degree)
    p = np.poly1d(coefs)
    ax.plot(x, p(x), color='red', label=f'Fit (degree {degree})')
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid(True)

# PbTe plots
plot_with_fit(axs[0, 0], pbte_data['temperature(K)'], pbte_data['electrical_conductivity(S/m)'], 2,
              'PbTe: Electrical Conductivity', 'Temperature (K)', 'Conductivity (S/m)')

plot_with_fit(axs[0, 1], pbte_data['temperature(K)'], pbte_data['seebeck_coefficient(μV/K)'], 2,
              'PbTe: Seebeck Coefficient', 'Temperature (K)', 'Seebeck Coefficient (μV/K)')

plot_with_fit(axs[0, 2], pbte_data['temperature(K)'], pbte_data['thermal_conductivity(W/mK)'], 3,
              'PbTe: Thermal Conductivity', 'Temperature (K)', 'Thermal Conductivity (W/mK)')

# SnSe plots
plot_with_fit(axs[1, 0], snse_data['temperature(K)'], snse_data['electrical_conductivity(S/m)'], 2,
              'SnSe: Electrical Conductivity', 'Temperature (K)', 'Conductivity (S/m)')

plot_with_fit(axs[1, 1], snse_data['temperature(K)'], snse_data['seebeck_coefficient(μV/K)'], 2,
              'SnSe: Seebeck Coefficient', 'Temperature (K)', 'Seebeck Coefficient (μV/K)')

plot_with_fit(axs[1, 2], snse_data['temperature(K)'], snse_data['thermal_conductivity(W/mK)'], 3,
              'SnSe: Thermal Conductivity', 'Temperature (K)', 'Thermal Conductivity (W/mK)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Load dataset
data = pd.read_excel('estm.xlsx')

# Define materials and references
pbte_data = data[data['Formula'] == 'PbTe']
snse_ref = 'https://doi.org/10.1016/j.jallcom.2016.01.190'
snse_data = data[(data['Formula'] == 'SnSe') & (data['reference'] == snse_ref)]

fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# Store fit equations and statistics
fit_results = []

# Enhanced plotting function with sorted polynomial fitting and statistical info
def plot_with_fit(ax, x, y, degree, title, xlabel, ylabel, material, prop_name):
    sorted_indices = np.argsort(x)
    x_sorted, y_sorted = x.iloc[sorted_indices], y.iloc[sorted_indices]

    coefs = np.polyfit(x_sorted, y_sorted, degree)
    p = np.poly1d(coefs)
    y_fit = p(x_sorted)
    
    ax.scatter(x_sorted, y_sorted, label='Data')
    ax.plot(x_sorted, y_fit, color='red', label=f'Fit (degree {degree})')
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid(True)

    r2 = r2_score(y_sorted, y_fit)
    fit_results.append({
        'Material': material,
        'Property': prop_name,
        'Polynomial': p,
        'R²': r2
    })

# PbTe plots
plot_with_fit(axs[0, 0], pbte_data['temperature(K)'], pbte_data['electrical_conductivity(S/m)'], 2,
              'PbTe: Electrical Conductivity', 'Temperature (K)', 'Conductivity (S/m)', 'PbTe', 'Electrical Conductivity')

plot_with_fit(axs[0, 1], pbte_data['temperature(K)'], pbte_data['seebeck_coefficient(μV/K)'], 2,
              'PbTe: Seebeck Coefficient', 'Temperature (K)', 'Seebeck Coefficient (μV/K)', 'PbTe', 'Seebeck Coefficient')

plot_with_fit(axs[0, 2], pbte_data['temperature(K)'], pbte_data['thermal_conductivity(W/mK)'], 3,
              'PbTe: Thermal Conductivity', 'Temperature (K)', 'Thermal Conductivity (W/mK)', 'PbTe', 'Thermal Conductivity')

# SnSe plots
plot_with_fit(axs[1, 0], snse_data['temperature(K)'], snse_data['electrical_conductivity(S/m)'], 2,
              'SnSe: Electrical Conductivity', 'Temperature (K)', 'Conductivity (S/m)', 'SnSe', 'Electrical Conductivity')

plot_with_fit(axs[1, 1], snse_data['temperature(K)'], snse_data['seebeck_coefficient(μV/K)'], 2,
              'SnSe: Seebeck Coefficient', 'Temperature (K)', 'Seebeck Coefficient (μV/K)', 'SnSe', 'Seebeck Coefficient')

plot_with_fit(axs[1, 2], snse_data['temperature(K)'], snse_data['thermal_conductivity(W/mK)'], 3,
              'SnSe: Thermal Conductivity', 'Temperature (K)', 'Thermal Conductivity (W/mK)', 'SnSe', 'Thermal Conductivity')

plt.tight_layout()
plt.show()

# Display polynomial fits and statistics
fit_df = pd.DataFrame(fit_results)
for idx, row in fit_df.iterrows():
    print(f"Material: {row['Material']} | Property: {row['Property']}\nPolynomial: {row['Polynomial']}\nR²: {row['R²']:.4f}\n")
    
# Save the polynomial dataframe as a CSV
fit_df.to_csv('polynomial_fits.csv', index=False)

# Save the figure as a PNG
fig.savefig('polynomial_fits.png', dpi=300)
    
```
